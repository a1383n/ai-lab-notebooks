```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np

X, y_true = make_blobs(n_samples=600, centers=3, cluster_std=1.3, random_state=42)

plt.scatter(X[:, 0], X[:, 1], s=20)
plt.title("Raw Data (make_blobs)")
plt.show()
```

```python
from sklearn.cluster import KMeans

model_sklearn = KMeans(n_clusters=3, n_init="auto", random_state=0)
labels_sklearn = model_sklearn.fit_predict(X)
centroids_sklearn = model_sklearn.cluster_centers_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels_sklearn, cmap="viridis", s=25)
plt.scatter(centroids_sklearn[:, 0], centroids_sklearn[:, 1], 
            c="red", marker="X", s=200)
plt.title("Sklearn KMeans - Result")
plt.show()
```

```python
x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1

xx, yy = np.meshgrid(np.linspace(x_min, x_max, 400),
                     np.linspace(y_min, y_max, 400))

grid = np.c_[xx.ravel(), yy.ravel()]

Z = model_sklearn.predict(grid).reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap="viridis")
plt.scatter(X[:, 0], X[:, 1], c=labels_sklearn, cmap="viridis", s=20)
plt.scatter(centroids_sklearn[:, 0], centroids_sklearn[:, 1], 
            c="red", marker="X", s=200)
plt.title("Sklearn KMeans - Decision Boundary")
plt.show()
```
